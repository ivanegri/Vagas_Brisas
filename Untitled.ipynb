```python
import pandas as pd
```

```python
df = pd.read_csv('brisas_bosque_itirapina_vagas.csv')
```

```python
df
```

```python
df['Unidade'] = df['Torre'].astype(str) + '-' + df['Apto'].astype(str)
```

```python
df.drop(columns={'Torre','Apto'}, inplace=True)
```

```python
df_pivot = df.pivot(index='Unidade',columns='Ano',values='Tipo da vaga').reset_index()
```

```python
df_pivot
```

```python
base = pd.read_csv('relacao_de_vagas.csv')
```

```python
base
```

```python
base_completa = pd.merge(base, df_pivot, on='Unidade')
```

```python
base_completa
```
